```python
"""1.Write a Python program that prints numbers from 1 to 5.
Expected Output:
1 2 3 4 5
"""

result_list:list = [val for val in range(1,6)]
for val in result_list:
  print(val, end=" ")
```

```python
"""2.Write a Python program to print the first 5 even numbers. Expected Output:
2 4 6 8 10"""

input:int = 5
result_list:list = [val for val in range(1,input*2+1) if val%2==0]
for val in result_list:
  print(val, end=" ")
```

```python
"""3.Write a Python program to print the sum of numbers from 1 to 10.
Expected Output:55"""

input:int = 10
result_list:list = [val for val in range(1,input+1)]
print(sum(result_list))

# Another way
result: int = 0
for val in result_list:
  result+=val
print(result)
```

```python
"""4.Write a Python function to find the factorial of 5. Expected Output:
120"""

input: int = 5
result: int = 1
for val in range(1, input+1):
  result*=val
print(result)
```
