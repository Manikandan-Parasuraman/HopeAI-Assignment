```python
def determine_the_amount(amount: float) -> str:
  """
  Determines whether the given amount is positive, negative, or zero.

  :param amount: amount is ready
  :type amount: float
  :return: Positive, Negative, or Zero
  :rtype: str
  """
  if amount > 0:
    return "Positive"
  elif amount < 0:
    return "Negative"
  else:
    return "Zero"

get_input: str = input("Enter the amount: ")
value: float = float(get_input)
result: str = determine_the_amount(amount=value)
print(f"The amount is {result}")
```

```python
def compute_the_sum_of_digits(number: int) -> int:
  """
  Computes the sum of the digits of a given number.

  :param number:
  """
  sum_of_digits: int = 0
  while number > 0:
    digit: int = number % 10
    sum_of_digits += digit
    number //= 10
  return sum_of_digits

get_input: str = input("Enter the number: ")
value: int = int(get_input)
result: int = compute_the_sum_of_digits(number=value)
print(f"The sum of the digits is {result}")
```

```python
def take_a_number_and_return_its_reverse(number: int) -> int:
  """
  takes a number and returns its reverse.

  :param number:
  :type number: int
  :return:
  :rtype: int
  """
  reverse: int = 0
  while number > 0:
      digit: int = number % 10
      reverse = reverse * 10 + digit
      number //= 10
  return reverse

get_input: str = input("Enter the number: ")
value: int = int(get_input)
result: int = take_a_number_and_return_its_reverse(number=value)
print(f"The reverse of the number is {result}")
```

```python
def check_if_a_given_number_is_prime(number: int) -> bool:
  """
  checks if a given number is prime.

  :param number:
  :type number: int
  :return:
  :rtype: bool
  """
  num: int = 2
  if number % num == 0:
    return False
  else:
    return True

get_input: str = input("Enter the number: ")
value: int = int(get_input)
result: bool = check_if_a_given_number_is_prime(number=value)
print(f"The number is prime: {result}")
```

```python
def find_the_factorial_of_a_given_number_using_recursion(number: int) -> int:
  """
  finds the factorial of a given number using recursion.

  :param number:
  :type number: int
  :return:
  :rtype: int
  """
  if number == 0 or number == 1:
    return 1
  else:
    return number * find_the_factorial_of_a_given_number_using_recursion(number - 1)

get_input: str = input("Enter the number: ")
value: int = int(get_input)
result: int = find_the_factorial_of_a_given_number_using_recursion(number=value)
print(f"The factorial of the number is {result}")
```

```python
def check_if_a_given_number_is_an_armstrong_number(number: int) -> bool:
  """
  Checks whether a given number is an Armstrong number.

  :param number: The number to check.
  :type number: int
  :return: True if the number is an Armstrong number, False otherwise.
  :rtype: bool
  """
  num_str = str(number)
  num_digits = len(num_str)
  sum_of_powers = 0
  for digit_char in num_str:
    sum_of_powers += int(digit_char) ** num_digits
  return sum_of_powers == number

get_input: str = input("Enter the number: ")
value: int = int(get_input)
result: bool = check_if_a_given_number_is_an_armstrong_number(number=value)
print(f"The number is an Armstrong number: {result}")
```

```python
def swap_first_and_last_characters(input_string: str) -> str:
  """
  Swaps the first and last characters of a given string.

  :param input_string: The string to modify.
  :type input_string: str
  :return: The string with the first and last characters swapped.
  :rtype: str
  """
  if len(input_string) < 2:
    return input_string

  first_char = input_string[0]
  last_char = input_string[-1]
  middle_part = input_string[1:-1]

  return last_char + middle_part + first_char

get_input: str = input("Enter the string: ")
value: str = get_input
result: str = swap_first_and_last_characters(input_string=value)
print(f"The string with swapped characters is: {result}")
```

```python
def convert_decimal_to_binary(decimal_number: int) -> str:
  """
  Converts a given decimal number into its binary equivalent.

  :param decimal_number: The decimal number to convert.
  :type decimal_number: int
  :return: The binary representation as a string.
  :rtype: str
  """
  if decimal_number == 0:
    return "0"

  binary_string = ""
  temp_number = decimal_number
  while temp_number > 0:
    remainder = temp_number % 2
    binary_string = str(remainder) + binary_string
    temp_number //= 2
  return binary_string

get_input: str = input("Enter the decimal number: ")
value: int = int(get_input)
result: str = convert_decimal_to_binary(decimal_number=value)
print(f"The binary equivalent is: {result}")
```

```python
def find_the_longest_word(sentence: str) -> str:
  """
  Finds the longest word in a sentence.

  :param sentence: The input sentence.
  :type sentence: str
  :return: The longest word in the sentence.
  :rtype: str
  """
  words = sentence.split()
  if not words:
    return ""

  longest_word = ""
  for word in words:
    clean_word = ''.join(char for char in word if char.isalnum()) # Remove punctuation
    if len(clean_word) > len(longest_word):
      longest_word = clean_word
  return longest_word

get_input: str = input("Enter the sentence: ")
value: str = get_input
result: str = find_the_longest_word(sentence=value)
print(f"The longest word is: {result}")
```

```python
def check_if_two_given_strings_are_anagrams(string1: str, string2: str) -> bool:
  """
  Checks whether two given strings are anagrams.

  :param string1: The first string.
  :type string1: str
  :param string2: The second string.
  :type string2: str
  :return: True if the strings are anagrams, False otherwise.
  :rtype: bool
  """
  normalized_string1 = sorted(string1.replace(" ", "").lower())
  normalized_string2 = sorted(string2.replace(" ", "").lower())

  return normalized_string1 == normalized_string2

get_input1: str = input("Enter the first string: ")
get_input2: str = input("Enter the second string: ")

value1: str = get_input1
value2: str = get_input2

result: bool = check_if_two_given_strings_are_anagrams(string1=value1, string2=value2)
print(f"The strings are anagrams: {result}")
```
