```python
class SubfieldsInAI:
  """
  A class to manage and display subfields within Artificial Intelligence.
  """
  def __init__(self):
    """
    Initializes the SubfieldsInAI class with a predefined list of subfields.
    """
    self.subfields_list:list = ["Machine Learning", "Neural Networks", "Vision", "Robotics", "Speech Processing", "Natural Language Processing"]
  def Subfields(self):
    """
    Prints each subfield of AI stored in the `subfields_list`.
    """
    for field in self.subfields_list:
      print(field)

print("Sub-fields in AI are:")
obj:object = SubfieldsInAI()
obj.Subfields()
```

```python
def find_odd_or_even(input:int) -> str:
  """
  Checks if a given integer is odd or even.

  :param input: An integer number.
  :type input: int
  :raises TypeError: if input is not an integer
  :returns: A string indicating whether the number is odd or even.
  :rtype: str
  """
  if not isinstance(input, int):
    raise TypeError("Input must be an integer")
  if input % 2 == 0:
    return f"{input} is Even Number"
  else:
    return f"{input} is Odd Number"

get_input:int = int(input("Enter a number to find odd or even : "))
print(find_odd_or_even(get_input))
```

```python
class ElegiblityForMarriage:
  """
  A class to check the eligibility for marriage based on gender and age.
  """
  def marriage_elegibility_checker(self, gender: str, age: int) -> str:
    """
    Checks if an individual is eligible for marriage based on their gender and age.

    :param gender: The gender of the individual ('male' or 'female').
    :type gender: str
    :param age: The age of the individual.
    :type age: int
    :raises TypeError: if age is not an integer
    :returns: A string indicating 'ELIGIBLE' or 'NOT ELIGIBLE'.
    :rtype: str
    """
    if age >= 21:
      return "ELIGIBLE"
    else:
      return "NOT ELIGIBLE"

obj: object = ElegiblityForMarriage()
get_gender: str = input("Enter your gender : ")
if get_gender.lower() in ["male", "female", "third gender"]:
  pass
else:
  raise ValueError("Gender must be 'male', 'female', or 'third gender'")

get_age: int = int(input("Enter your age : "))
if not isinstance(get_age, int):
  raise TypeError("Age must be an integer")

if get_age < 0:
  raise ValueError("Age cannot be negative")

print(obj.marriage_elegibility_checker(get_gender, get_age))
```

```python
class FindPercent:
  """
  A class to calculate the total and percentage from a list of subject marks.
  """
  def percentage_calculator(self, subjects: list) -> tuple:
    """
    Calculates the total marks and the overall percentage from a list of subject scores.

    Assumes each mark in the `subjects` list is a score out of 100.

    :param subjects: A list of integer marks obtained in various subjects.
    :type subjects: list[int]
    :returns: A tuple containing the total marks and the calculated percentage.
    :rtype: tuple[int, float]
    """
    total = sum(subjects)
    percentage = total / len(subjects)
    return total, percentage

obj: object = FindPercent()

subject_list: list = ["Tamil", "English", "Math", "Science", "Social"]
marks: list = []
for subject in subject_list:
  while True:
    try:
      mark = int(input(f"Enter your {subject} marks : "))
      if 0 <= mark <= 100:
        marks.append(mark)
        break
      else:
        print("Mark must be between 0 and 100. Please try again.")
    except ValueError:
      print("Invalid input. Please enter an integer.")

total, percentage = obj.percentage_calculator(marks)
print(f"Total : {total}")
print(f"Percentage : {percentage}")
```

```python
class Triangle:
  """
  A class to calculate the area and perimeter of a triangle.
  """
  def calculate_area(self, height: float, breadth: float) -> float:
    """
    Calculates the area of a triangle given its height and breadth.

    :param height: The height of the triangle.
    :type height: float
    :param breadth: The breadth (base) of the triangle.
    :type breadth: float
    :raises ValueError: If height or breadth is not a positive number.
    :returns: The area of the triangle.
    :rtype: float
    """
    if not isinstance(height, (int, float)) or height <= 0:
      raise ValueError("Height must be a positive number.")
    if not isinstance(breadth, (int, float)) or breadth <= 0:
      raise ValueError("Breadth must be a positive number.")
    return (height * breadth) / 2

  def calculate_perimeter(self, side1: float, side2: float, side3: float) -> float:
    """
    Calculates the perimeter of a triangle given the lengths of its three sides.

    :param side1: The length of the first side.
    :type side1: float
    :param side2: The length of the second side.
    :type side2: float
    :param side3: The length of the third side.
    :type side3: float
    :raises ValueError: If any side is not a positive number.
    :returns: The perimeter of the triangle.
    :rtype: float
    """
    if not isinstance(side1, (int, float)) or side1 <= 0:
      raise ValueError("Side 1 must be a positive number.")
    if not isinstance(side2, (int, float)) or side2 <= 0:
      raise ValueError("Side 2 must be a positive number.")
    if not isinstance(side3, (int, float)) or side3 <= 0:
      raise ValueError("Side 3 must be a positive number.")
    return side1 + side2 + side3
```

```python
obj_triangle: object = Triangle()

# Get input for Area
height: float = 0
breadth: float = 0
while True:
  try:
    height = float(input("Height:"))
    if height <= 0:
      print("Height must be a positive number. Please try again.")
      continue
    breadth = float(input("Breadth:"))
    if breadth <= 0:
      print("Breadth must be a positive number. Please try again.")
      continue
    break
  except ValueError:
    print("Invalid input. Please enter a numerical value.")

# Calculate and print Area
area = obj_triangle.calculate_area(height, breadth)
print("Area formula: (Height*Breadth)/2")
print(f"Area of Triangle: {area}")

# Get input for Perimeter
side1 = 0
side2 = 0
side3 = 0
while True:
  try:
    side1: float = float(input("Height1:"))
    if side1 <= 0:
      print("Height1 (Side 1) must be a positive number. Please try again.")
      continue
    side2 = float(input("Height2:"))
    if side2 <= 0:
      print("Height2 (Side 2) must be a positive number. Please try again.")
      continue
    side3 = float(input("Breadth:"))
    if side3 <= 0:
      print("Breadth (Side 3) must be a positive number. Please try again.")
      continue
    break
  except ValueError:
    print("Invalid input. Please enter a numerical value.")

# Calculate and print Perimeter
perimeter = obj_triangle.calculate_perimeter(side1, side2, side3)
print("Perimeter formula: Height1+Height2+Breadth")
print(f"Perimeter of Triangle: {perimeter}")
```
