```python
"""
A system checks if a user is eligible to vote based on their age.
Write logic to ask the user for their age and determine if they are eligible to vote based on whether they are 18 or older.
"""

def vote_eligibility_checker(age: int) -> str:
    """
    Check whether a person is eligible to vote based on their age.

    : param age: Age of the person in years.
    : type age: int
    :return: Eligibility message.
    :rtype: str
    """
    if age >= 18:
        return "You are eligible to vote"
    else:
        return "You are not eligible to vote"


def validate_age(age: str) -> int | str:
    """
    Validate and convert the age input

    Attempts to convert the input string to an integer and
    validate that the age is non-negative.

    :param age: Age provided as a string input.
    :type age: str
    :return: validate age as a integer or an error message.
    :rtype: int | str
    """
    try:
        age_int = int(age)
    except ValueError:
        return "Age should be a number"

    if age_int < 0:
        return "Age should not be less than zero"

    return age_int


user_input: str = input("Enter your age: ")
data: int | str = validate_age(age=user_input)

if isinstance(data, int):
    print(vote_eligibility_checker(age=data))
else:
    print(data)
```

```python
"""
A program processes a list of numbers and needs to find the largest value.
Write logic to identify and return the largest number from a given list
"""
def find_largest_number(numbers: list[int]) -> int:
  """
  Find the largest number in a list of integers.

  :param numbers: list of integers.
  :type numbers: list[int]
  :return: largest number in the list.
  :rtype: int
  """
  largest = 0
  for num in numbers:
    if num > largest:
      largest = num
  return largest

input_numbers: str = input("Enter a list of numbers separated by spaces: ")
numbers: list = [int(num) for num in input_numbers.split()]
largest_number: int = find_largest_number(numbers)
print("The largest number is:", largest_number)
```

```python
"""
A company provides employees with a 10% bonus if their salary exceeds $50,000.
Write logic to determine the bonus amount based on the given salary
"""
def calculate_bonus(salary: float) -> float:
  """
  Calculate the bonus amount based on the given salary.

  Bonus will be the 10% of the given salary

  :param salary: salary of the employee.
  :type salary: float
  :return: bonus amount.
  :rtype: float
  """
  bonus:float = 0.10 * salary
  return bonus

salary: str = float(input("Enter your salary: "))
bonus: float = calculate_bonus(salary)
print("Your bonus is:", bonus)
```

```python
"""
A program evaluates a number to determine if it is even or odd.
Write logic to check whether a given number is even or odd.
"""
def find_odd_or_even(number: int) -> str:
  """
  Find whether a given number is even or odd.

  :param number: number to check
  :type number: int
  :return: even or odd
  :rtype: str
  """
  if number % 2 == 0:
    return f"Entered {number} is Even"
  else:
    return f"Entered {number} is Odd"

number:str = int(input("Enter a number: "))
result:int = find_odd_or_even(number)
print(result)
```

```python
"""
A text-processing tool reverses a given word or sentence for formatting purposes.
Write logic to take a word or sentence as input and produce its reversed version.
"""
def reverse_word(word: str) -> str:
  """
  Reverse a given word or sentence.

  :param word: word or sentence to reverse
  :type word: str
  :return: reversed word or sentence
  :rtype: str
  """
  return word[::-1]

word:str = input("Enter a word or sentence: ")
result:str = reverse_word(word)
print(result)
```

```python
"""
A grading system determines whether a student has passed or failed based on their score.
Write logic to check if a student has passed a subject by scoring at least 40 marks.
"""
def student_result(marks: int) -> str:
  """
  Check the result of a student

  :param marks: marks of the student
  :type marks: int
  :return: result of the student
  :rtype: str
  """
  if marks >= 40:
    return "Pass"
  else:
    return "Fail"

input_mark: str = input("Enter your marks: ")
marks: int = int(input_mark)
result:str = student_result(marks)
print(result)
```

```python
"""
A retail store offers a 20% discount if a customerâ€™s total order exceeds $100. Write logic to calculate the final amount to be paid after applying the discount.
"""
def calculate_final_amount(total_order: float) -> float:
  """
  Calculate the final amount to be paid after applying the discount.

  :param total_order: total order amount
  :type total_order: float
  :return: final amount to be paid
  :rtype: float
  """
  result = total_order - (total_order * 0.20)
  return result

input_total_order: str = input("Enter your total order amount: ")
total_order: float = float(input_total_order)
final_amount: float = calculate_final_amount(total_order)
print(final_amount)
```

```python
"""
A banking system processes withdrawal requests and ensures the user has enough balance.
"""
def check_balance(balance: float, withdrawal_amount: float) -> bool:
  """
  Check if the user has enough balance to withdraw the given amount.

  :param balance: current balance of the user
  :type balance: float
  :param withdrawal_amount: amount to withdraw
  :type withdrawal_amount: float
  :return: True if the user has enough balance, False otherwise
  :rtype: bool
  """
  if balance >= withdrawal_amount:
      return True
  else:
    return False

balance:float = float(input("Enter your current balance: "))
withdrawal_amount: float = float(input("Enter the amount to withdraw: "))
response: bool = check_balance(balance, withdrawal_amount)
if response:
  print("You have enough balance to withdraw the given amount")
else:
  print("You don't have enough balance to withdraw the given amount")
```

```python
"""
A calendar system verifies whether a given year is a leap year based on standard leap year rules.
Write logic to determine whether a given year is a leap year
"""
def check_leap_year(year: int) -> bool:
  """
  Check if a given year is a leap year.

  :param year: year to check
  :type year: int
  :return: True if the year is a leap year, False otherwise
  :rtype: bool
  """
  if year % 400 == 0:
      return True
  if year % 100 == 0:
      return False
  if year % 4 == 0:
      return True
  return False

get_input:str = input("Enter a year: ")
year:int = int(get_input)
result:bool = check_leap_year(year)
if result:
  print(f"{year} is a leap year")
else:
  print(f"{year} is not a leap year")
```

```python
"""
A program filters out only even numbers from a given list.
 Write logic to extract and return only the even numbers from a list.
"""
def filter_only_even_number(numbers: list[int]) -> list[int]:
  """
  Filter out only the even numbers

  :param numbers: list of numbers
  :type numbers: list[int]
  :return: list of even numbers
  :rtype: list of numbers
  """
  response_list = []
  for val in numbers:
    if val % 2 == 0:
      response_list.append(val)
  return response_list

input_numbers: str = input("Enter a list of numbers separated by spaces: ")
numbers_list: list[int] = [int(num) for num in input_numbers.split()]

even_numbers: list[int] = filter_only_even_number(numbers=numbers_list)
print(even_numbers)
```
