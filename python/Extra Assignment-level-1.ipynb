```python
get_value_list = ['name', 'age', 'school_name', 'degree']
data_dict = {}
for get_value in get_value_list:
  data_dict[get_value] = input(f"Enter your {get_value}: ")
print("*"*20)
for k, v in data_dict.items():
  print(f"{k.title()}: {v}" )
```

```python
value_a = int(input("Enter Value A: ",))
value_b = int(input("Enter Value B: ",))
print("*"*20)
print("Addition: 1")
print("Subtraction: 2")
print("MultiplicationL: 3")
print("Division: 4")
print("Floor Division: 5")
print("Modulo: 6")
print("Power: 7")
print("*"*20)
choice = int(input("Choose Options: ", ))
print("*"*20)
match choice:
  case 1:
      print("Addition: ", value_a+value_b)
  case 2:
      print("Subtraction: ", value_a-value_b)
  case 3:
      print("Multiplication: ", value_a*value_b)
  case 4:
      print("Division: ", value_a/value_b)
  case 5:
      print("Floor Division: ", value_a//value_b)
  case 6:
      print("Modulo: ", value_a%value_b)
  case 7:
      print("Power: ", value_a**value_b)
  case _:
      print("Invalid choice")
```
